## 5. Modelling

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
data_copy.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_copy.iloc[:,:], data_copy['target_class'], random_state = 42)
X_train_data = X_train.drop('target_class', axis=1)

target = X_train['target_class']

X_train_data.shape, target.shape
```

```python
target.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

### 6.2 Cross Validation (K-fold)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

### 6.2.1 kNN

```python
clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, X_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```

### 6.2.2 Decision Tree

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, X_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# decision tree Score
round(np.mean(score)*100, 2)
```

### 6.2.3 Random Forest

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, X_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score)*100, 2)
```

### 6.2.4 Naive Bayes

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, X_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Naive Bayes Score
round(np.mean(score)*100, 2)
```

### 6.2.5 SVM

```python
clf = SVC(gamma='scale')
scoring = 'accuracy'
score = cross_val_score(clf, X_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```
