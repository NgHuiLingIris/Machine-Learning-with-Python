# Scatter Plot and Correlation

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
dataset = pd.read_csv('pulsar_stars.csv').dropna()
```

```python
dataset.columns = ['Mean IP','SD IP', 'EK IP','Skew IP','Mean DM-SNR','SD DM-SNR', 'EK DM-SNR','Skew DM-SNR','target']
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(dataset.iloc[:,0:7], alpha=0.2, figsize=(10, 10), diagonal='target_class')
```

```python
import seaborn as sns
sns.set(style="ticks")

sns.pairplot(dataset, hue="target")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

cax = ax1.imshow(dataset.corr(), interpolation="nearest")
ax1.grid(True)
plt.title('Correlation ')
labels=['Mean IP','SD IP', 'EK IP','Skew IP','Mean DM-SNR','SD DM-SNR', 'EK DM-SNR','Skew DM-SNR']
ax1.set_xticklabels(labels,fontsize=6)
ax1.set_yticklabels(labels,fontsize=6)
# Add colorbar, make sure to specify tick locations to match desired ticklabels
fig.colorbar(cax, ticks=[-0.4,0,.4,.6,.8,.9,1])
plt.show()
```
